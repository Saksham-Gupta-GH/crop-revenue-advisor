```python
import pandas as pd

crop = pd.read_csv("data/crop_production.csv")

print("SHAPE:", crop.shape)
print("\nCOLUMNS:")
print(crop.columns.tolist())

print("\nSAMPLE ROWS:")
print(crop.head(5))

print("\nUNIQUE STATES:", crop["State_Name"].nunique())
print("UNIQUE DISTRICTS:", crop["District_Name"].nunique())
print("UNIQUE CROPS:", crop["Crop"].nunique())

print("\nUNIQUE SEASONS:")
print(crop["Season"].unique())
```

```python
rain = pd.read_csv("data/rainfall_district.csv")

print("SHAPE:", rain.shape)
print("\nCOLUMNS:")
print(rain.columns.tolist())

print("\nSAMPLE ROWS:")
print(rain.head(5))

print("\nUNIQUE STATES:", rain["STATE_UT_NAME"].nunique())
print("UNIQUE DISTRICTS:", rain["DISTRICT"].nunique())

print("\nANNUAL RAINFALL STATS:")
print(rain["ANNUAL"].describe())
```
